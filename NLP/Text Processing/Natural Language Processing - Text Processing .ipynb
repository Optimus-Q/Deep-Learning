### Text Processing Steps:

* Get all the letters into one format lower case or upper case

* tokenize words using word tokenize method

* remove all punctuations

* remove stopword 

* Stemming or lemmatization


```python
# Get all the letters into one format lower case or upper case
```

```python
lower_case = text.lower()   #lower case all the letters
```

```python
text   #actual text
```

```python
lower_case   # lower case text
```

```python
# I am using tokenize words now, because it will be easier to do other text processing methods (individual preference)
# In tokenize you get text splitted into each words in a list
token_words = word_tokenize(lower_case)
token_words
```

```python
len(token_words)   # total tokens
```

Now those tokens contains punuctuations, therefore we create our function to remove them

```python
#function takes each tokn words and checks if each words contains punctuations and if exists it replaces with empty string and
#appends in "word join" list, after all token words the function returns the clean words.

def Punctuator(words):
    punctuations = '''~`@#$%^&*()_-+=/*,-.[]""{};:’''""“” <>¢?\—|1234567890\n'''
    word_join = []
    for t in words:
        for p in punctuations:
            if (p in t):
                t = t.replace(p, '')
        word_join.append(t)
    return(word_join)
```

```python
punctuated_text = ' '.join(Punctuator(token_words))
punctuated_text
```

Now steps like, stopwords removal and stemming and lemmatization is individual preference, because these methods removes the required information for applications such as text generation.

However, for sentiment analysis we can do the above steps.

```python
# stopwords remove

print(stopwords.words('english'))   #some of the usual words, which we will remove from the text
```

```python
punctuated_words = Punctuator(token_words)
stopwords_text = [i for i in punctuated_words if i not in stopwords.words('english')]
```

```python
punctuated_words[:10]
```

```python
stopwords_text # we can see that words like "i, am, to, be, with, you" removed
```

```python
# stemming and lemmatization

# stemming reduces the words into its root form that is[go,going] into "go"
ps = PorterStemmer()
ps_words = []   
for w in (stopwords_text):
    ps_words.append(ps.stem(w))
```

```python
ps_words[:10]    #we can see that some of the end letters are removed using stemming method
```

### Stemming vs Lemmatization

```python
qw = ['like', 'liked', 'liking', 'go', 'going', 'god']
stm = PorterStemmer()
stm_words = []   
for w in (qw):
    stm_words.append(stm.stem(w))
```

```python
stm_words
```

```python
qw = ['like', 'liked', 'liking', 'likes', 'go', 'going', 'god']
lem = WordNetLemmatizer()
lem_words = []   
for w in (qw):
    lem_words.append(lem.lemmatize(w))
```

```python
lem_words
```

we can see that input for stemming and lemmatizing were same but the outputs were much meaningful in lemmatization than stemming, therefore many people prefer lemmatization because it given better meanningful words after words normalization.

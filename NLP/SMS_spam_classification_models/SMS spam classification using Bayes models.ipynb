```python
# Import libaries

import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
# Import Data

sms = pd.read_csv("SMSSpamCollection.csv", 
                 sep = '\t', names = ['label', 'message'])
sms
```

```python
sms_text = sms['message']
sms_text
```

```python
#this is my text cleaner, which lets me to clean the entire text  (it is not efficient enough now, more methods need to introduce to
#do better cleaning)


import TextCleaner
```

```python
tc = TextCleaner.TextCleaner(text = sms_text,
                             text_form = 'List',
                             stemming = True,
                             lemmat = False)
```

```python
processed_text = tc.clean()
```

```python
processed_text[:4]
```

```python
#words to vector formarion

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf =  TfidfVectorizer()
X_tfidf = tfidf.fit_transform(processed_text).toarray()
```

```python
X_tfidf
```

```python
tfidf.get_feature_names()
```

```python
Y = np.where(sms['label'] == 'ham', 0, 1).reshape(-1, 1)   # Label data
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X_tfidf,Y, train_size = 0.7, shuffle = True)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(xtrain, ytrain)
pred = clf.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,pred)   #overall predicted accuarcay
```

```python
#some times predicted class will be same every time so in order to check that biasness we use each class accuracy
#Distribution of test data doesnt matter, it only matters in training therefore we need to have enough both class in training to predict better in test.

def EachAccuracyClass(yt, pt):
    cls, ncls = np.unique(yt, return_counts = True)
    acc = pd.DataFrame()
    acc['ytest'] = yt.squeeze()
    acc['pred'] = pt.squeeze()
    class_acc = {}
    for c in cls:
        sig = acc.loc[acc.loc[acc['ytest']==c].index]
        class_acc[str(c)] = round(accuracy_score(sig['ytest'], sig['pred'])*100, 2)
    return (class_acc)
```

```python
EachAccuracyClass(ytest, pred)
```
